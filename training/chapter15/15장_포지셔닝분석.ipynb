## 15.1 포지셔닝 분석 기초
- 포지셔닝 분석
  - __분석 대상의 개체를 좌표평면 상__에 포지셔닝 시키는 분석기법
  - 두 개의 축 (두 가지 변수)에 대한 측정값을 대응시켜 표시하는 일반적인 __맵핑(Mapping)__도 넓은 의미의 포지셔닝 분석
  - __요인분석, 군집분석__과 같은 차원축소 또는 그룹핑 방법 역시 포지셔닝 기법으로 활용 가능
    - 요인분석, 군집분석은 원천적으로 포지셔닝을 위한 분석도구가 아님, __전문적인 포지셔닝 분석에는 부적합__
  - 그러나 __변수간의 중요도 반영, 차원 증가__ 등의 경우 사용 불가능

- 마케팅 통계분석 기법 (기업이나 상품, 브랜드 같은 개체들의 포지셔닝을 전문적으로 수행하는 분석)
  - __다차원 척도법__ (MDS: Multi-dimensional Scaling)
    - __등간, 비율척도__로 측정된 데이터를 바탕으로 포지셔닝 수행
  - __상응분석__ (Correspondence Analysis)
    - 주로 __명목척도__로 측정된 데이터를 바탕으로 포지셔닝 수행
  - 공통점 : 차원축소로 2차원 좌표평면과 같이 시각적으로 판단 용이한 방법으로 개체들을 표현

## 15.2 다차원 척도법

### [1] 개념 및 주요 이론

- 다차원 척도법 (MDS)

  - 브랜드, 기업 등과 같이 __비교 대상 개체__들의 __다양한 속성 값__을 바탕으로 다차원 관계를 __저차원(차원축소) 2차원, 3차원__에 단순한 구도로 시각화하여 표현하는 분석기법

  - 고객의 만족도나 선호도 측정, 이를 바탕으로 브랜드간의 심리적 거리를 계산하여 인지도를 표현하는데 자주 사용
    - 인지도 분석이라고도 불림

  - 복잡한 __분석결과를 해석할 필요 없고, 시각적으로 직관적으로 이해할 수 있다는 장점__

image.png

- 위는 다차원 척도법을 통해 __6가지 골프클럽 브랜드에 대한 경쟁구도__ 확인 가능
  - 상표 A와 B는 직접적인 경쟁관계
  - 그러나 A와 D는 약한 경쟁관계

- 다차원 척도법 결과는 __주요 속성들간의 연관성__ 또한 확인 가능
  - 방향성과 비거리라는 속성은 강하게 연관되어 있음

#### 다차원 척도법의 개념적 분석과정
- 크게 4가지의 단계를 거쳐 수행

##### ① 자극점에 대한 속성과 평가 결과 요약
- 필요 데이터 준비
- 자극점에 대한 주요 속성 평가값을 요약 및 요약 자료 확보
  - 아래 표 예시는 5개 주요 매장(자극점)에 대한 매출, 재고량 등 6가지 변수(속성)들의 측정값들을 매장 별로 요약

image.png

##### ② 속성간의 거리행렬 도출
- 각 속성들의 평균 점수를 바탕으로 속성들간 평균 거리행렬을 도출
- 속성간의 거리는 일반적으로 유클리디안 거리 척도를 사용
  - 유클리디안 거리 척도 : 각 측정변수들의 차이를 제곱한 값들을 모두 합한 후 제곱근을 취한 값, 두 개체들 간의 최소거리 제공

##### ③ 차원 좌표 생성
- 평균거리행렬 상에 제시된 개체간의 거리 값을 바탕으로 좌표 값을 생성

image.png

##### ④ 좌표계 맵핑
- 좌표 값을 바탕으로 좌표계에 자극점, 속성을 표기
- 맵에 나타나는 각 축은 특정 속성이 아닌, 단지 분류 차원을 구분하기 위한 목적
  - 각 축에 대해 특별한 의미를 부여하는 것은 부적절

#### 차원 수 선정 방법
- 분석 대상 개체들의 관계를 적절한 차원의 수로 표현해야 함
- 요인분석과 유사 (다양한 속성을 적은 수의 차원으로 요약하여 분석)
- 적절한 차원의 수를 결정하는 방법
  - 크루스칼(Kruskal)의 스트레스(stress)값을 사용
    - 요인분석에서도 동일 목적으로 사용 가능
    - 다차원 정보를 저차원으로 축약하는 과정에서 발생하는 __정보의 손실율의 개념__
    - __측정에 의해 나타난 실제 평가 거리__와 __다차원 척도법에 의해 추정된 거리__를 바탕으로 계산

image.png

- 크루스칼의 스트레스
  - 0과 1사이의 값을 가짐
    - 0에 가까울수록 적합도 높음, 1에 가까울수록 적합도 낮음
    - 일반적으로 0.1 이하일 경우 차원 수가 적절하다고 판단
  - 차원 수를 늘려가며 스트레스 값의 변화를 추적, 가장 크게 감소하는 차원의 수를 찾아 최적의 수로 결정

image.png

- 최적의 수를 결정 하였다고 하더라도 3차원 이상일 경우 표현에 한계가 있음
  - 대부분 2차원 또는 3차원이 바람직
- 따라서 스트레스 값은 __차원을 축소하는 결정에 대한 근거나 타당성을 검토하기 위한 보조적인 역할로 활용__

#### 단순 경쟁구조 분석과정
- A쇼핑의 포지셔닝 분석을 위해 6개 업체와 단순 경쟁구조 분석을 수행하고자 함
- 자극점은 7개 쇼핑몰, 속성은 이미지, 접근성 등 7개로 설정

- 자극점 차원 좌표 생성을 위한 MDS() 함수
  - 표현 방식 : MDS(n_components=차원 수, random_state=난수)
    - 2차원 설정과 동일한 결과값을 보기 위해 난수 123으로 설정
  - fit()함수로 이미지:가격만족도의 데이터를 적합
  - fit_transform()속성으로 차원 좌표를 X_mds에 저장

#### 속성-경쟁구조 분석과정
- 경쟁구조와 속성을 동시에 확인하기 위한 포지셔닝 실행

```
# 1. 모듈 및 데이터 탑재
import pandas as pd
import statsmodels.formula.api as smf  # 회귀 분석을 위한 패키지 호출
from sklearn.manifold import MDS
import numpy as np

df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/statistics/data/MDS1.csv',sep=',',encoding='CP949')
df1 = df.loc[:,'이미지':'가격만족도']
df1 = (df1-df1.min())/(df1.max()-df1.min())  # 정규화(요소값-최소값)/(최대값-최소값) : 동일한 척도로 계산하기 위함
df1.head()
```

```
# 2. 다차원척도법 분석 (차원 좌표 값 계산)
clf = MDS(n_components=2, random_state=123).fit(df1)
X_mds = clf.fit_transform(df1)
X_mds
```

```
# 3. 속성 차원 좌표 값 계산
df1['차원1']=X_mds[:,0]
df1['차원2']=X_mds[:,1]

model=[]
model.append(smf.ols(formula='이미지~차원1+차원2',data=df1).fit())  # 속성 : 종속변수
model.append(smf.ols(formula='접근성~차원1+차원2',data=df1).fit())  # 자극점의 차원 좌표 값 : 독립변수
model.append(smf.ols(formula='서비스~차원1+차원2',data=df1).fit())  # 회귀분석 수행
model.append(smf.ols(formula='친절성~차원1+차원2',data=df1).fit())
model.append(smf.ols(formula='편의시설~차원1+차원2',data=df1).fit())
model.append(smf.ols(formula='인지도~차원1+차원2',data=df1).fit())
model.append(smf.ols(formula='가격만족도~차원1+차원2',data=df1).fit())

속성=[]  # 독립변수의 회귀계수를 다시 좌표값으로 하여 속성 리스트에 부여
for i in range(0,7,1):
  속성.append([model[i].params[1],model[i].params[2]])
속성 = np.array(속성)
자극점및속성=np.hstack([X_mds,속성]) # 자극점의 차원 좌표, 속성의 차원 좌표를 붙임
자극점및속성
```

- 속성-경쟁구조 분석 시각화

```
# 1. 속성-경쟁구조 시각화
labels = df.shop
for label, x_count, y_count in zip(labels, 자극점및속성[:,0], 자극점및속성[:,1]):
  plt.annotate(label,
               xycoords='data',
               textcoords='offset points',
               xy=(x_count, y_count),
               xytext=(5,-5))

labels2 = df1.columns
for label, x_count, y_count in zip(labels2, 자극점및속성[:,2], 자극점및속성[:,3]):
  plt.annotate(label,
               xycoords='data',
               textcoords='offset points',
               xy=(x_count, y_count),
               xytext=(5,-5))
  
plt.xlabel('Dimension1')
plt.ylabel('Dimension2')
plt.scatter(자극점및속성[:,0],자극점및속성[:,1]) # 자극점 좌표 그래프
plt.scatter(자극점및속성[:,2],자극점및속성[:,3]) # 속성 좌표 그래프
plt.show()
```

- A 쇼핑몰의 경우 친절성에 상당히 경쟁력이 있음
  - 반면, 이미지, 접근성 관점에서는 다소 경쟁력이 약함
- D, F 쇼핑몰은 기업 이미지와 접근성 관점에서 매우 강력한 경쟁관계임을 알 수 있음

#### 이상점-경쟁구조 분석과정
- 기업간의 경쟁구조에 고객들의 선호위치도 함께 표현하고자 함

```
# 1. 모듈 및 데이터 탑재
df1 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/statistics/data/MDS1.csv',sep=',',encoding='CP949')
df3 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/statistics/data/MDS3.csv',sep=',',encoding='CP949')

df3.head()
```

```
# 2. 다차원척도법 분석(좌표 값 계산)
clf = MDS(n_components=2, random_state=123)
X_mds1 = clf.fit_transform(df1.loc[:,'이미지':'가격만족도'])
X_mds3 = clf.fit_transform(df3.loc[:,'A쇼핑':'G쇼핑'])

print('자극점 차원좌표\n',X_mds1,'\n')
print('이상점 차원좌표\n',X_mds3[:5],'\n')
print('이상점 차원좌표 수\n',len(X_mds3))
```

- 이상점-경쟁구조 시각화

```
# 1. 이상점-경쟁구조 시각화
labels = df1.shop
for label, x_count, y_count in zip(labels, X_mds1[:,0], X_mds1[:,1]):
  plt.annotate(label,
               xycoords='data',
               textcoords='offset points',
               xy=(x_count, y_count),
               xytext=(5,-5))
plt.xlabel('Dimension1')
plt.ylabel('Dimension2')
plt.scatter(X_mds1[:,0],X_mds1[:,1],label='자극점')
plt.scatter(X_mds3[:,0],X_mds3[:,1],label='이상점')
plt.scatter(X_mds3[:,0].mean(),X_mds3[:,1].mean(),label='평균 이상점')
plt.legend()
plt.show()
```

- 고객들의 평균 이상점 주변에는 B 쇼핑몰이 가장 가까운 곳에 위치
- D, F 쇼핑몰은 경쟁관계이나, 고객들의 이상점이 많지 않음
  - 평균 이상점으로부터도 상당히 멀리 떨어져 있음
  - 시장지배력은 약할 것이라고 판단
