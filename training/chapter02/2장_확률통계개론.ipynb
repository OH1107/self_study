### [1] 확률

- 확률이란 경험 혹은 실험의 결과로 특정한 사건(event)이나 결과가 발생할 가능성
  - 주사위를 던져서 1이 나올 가능성이 1/6
  - 일기 예보에서 비가 올 가능성이 30%라고 하는 것처럼 사건이 일어날 수 있는 가능성의 정도를 확률이라고 함

#### 1) 확률의 정의
- 확률의 고정적 정의는 등확률 모형에 입각하여 정의된 것
  - 표본공간이 N개의 원소로 이뤄져 있고,
  - 각 원소가 동등한 확률로 일어난다고 가정할 때,
  - 사건 A에 해당하는 원소의 개수가 n(A)라고 하면 사건 A의 확률을 n(A)/N으로 정의
- 등확률 가정이 현실적이지 못하기 때문에 상대도수의 극한 개념으로서 통계적 확률을 정의

#### 3) 조건부 확률

- 사상 A가 일어났다는 조건하에 사상 B가 일어날 확률을 조건부 확률이라고 함
  - 이를 P(B|A)로 표현

image.png

- 조건부 확률은 개념적으로 두 사상의 독립을 정의할 때 유용
- 만일 사상 A는 사상 B가 일어나는 확률에 영향을 미치지 않는다면
  - 두 사상 A와 B가 독립이라는 것을 뜻함

image.png

- 독립사건과 배반사건은 서로 다른 의미라는 것을 유의해야 함
- 배반사건 : 두 사건이 동시에 일어날 확률이 0인 경우
- 독립사건 : 두 사건이 동시에 일어날 확률이 각각의 사건이 일어날 확률의 곱으로 주어지는 경우
